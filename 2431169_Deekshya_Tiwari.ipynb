Task - 3 - Outlier Detection:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the South Asian dataset
south_asia_data = pd.read_csv('south_asian_countries.csv')

# Step 1: Identify outliers using the 1.5  IQR rule for 'Score' and 'GDP per Capita'

# Function to calculate IQR and identify outliers
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Identify outliers for 'Score' and 'GDP per Capita'
outliers_score = identify_outliers(south_asia_data, 'score')
outliers_gdp = identify_outliers(south_asia_data, 'Log GDP per capita')

# Step 2: Create a scatter plot and highlight outliers
plt.figure(figsize=(10, 6))

# Scatter plot for GDP per Capita vs Score
plt.scatter(south_asia_data['Log GDP per capita'], south_asia_data['score'], color='blue', label='Normal Data')

# Highlight outliers in a different color (red)
plt.scatter(outliers_score['Log GDP per capita'], outliers_score['score'], color='red', label='Outliers (score)', s=100, edgecolor='black')

# Highlight outliers in GDP per Capita in a different color (green)
plt.scatter(outliers_gdp['Log GDP per capita'], outliers_gdp['score'], color='green', label='Outliers (Log GDP per capita)', s=100, edgecolor='black')

# Add labels and title
plt.title('Scatter Plot: GDP per Capita vs Happiness Score with Outliers', fontsize=16)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 3: Discuss the characteristics of outliers and their potential impact on regional averages

# Display outliers
print("Outliers based on Score:")
print(outliers_score[['Country name', 'score', 'Log GDP per capita']])

print("\nOutliers based on GDP per Capita:")
print(outliers_gdp[['Country name', 'score', 'Log GDP per capita']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load the South Asian dataset
south_asia_data = pd.read_csv('south_asian_countries.csv')

# Choose two metrics: 'Freedom to Make Life Choices' and 'Generosity'
metric_1 = 'Freedom to make life choices'
metric_2 = 'Generosity'

# Calculate Pearson correlation between the metrics and the Score
correlation_1, _ = pearsonr(south_asia_data[metric_1], south_asia_data['score'])
correlation_2, _ = pearsonr(south_asia_data[metric_2], south_asia_data['score'])

print(f"Pearson correlation between {metric_1} and Score: {correlation_1:.2f}")
print(f"Pearson correlation between {metric_2} and Score: {correlation_2:.2f}")

# Task 2: Create scatter plots with trendlines
plt.figure(figsize=(12, 6))

# Scatter plot for 'Freedom to Make Life Choices' vs Score
plt.subplot(1, 2, 1)
sns.regplot(x=metric_1, y='score', data=south_asia_data, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title(f'Scatter Plot: {metric_1} vs Happiness Score')
plt.xlabel(metric_1)
plt.ylabel('Happiness Score')

# Scatter plot for 'Generosity' vs Score
plt.subplot(1, 2, 2)
sns.regplot(x=metric_2, y='score', data=south_asia_data, scatter_kws={'color': 'green'}, line_kws={'color': 'red'})
plt.title(f'Scatter Plot: {metric_2} vs Happiness Score')
plt.xlabel(metric_2)
plt.ylabel('Happiness Score')

plt.tight_layout()
plt.show()

# Task 3: Discuss Relationships
# Based on the correlation values and the scatter plots, we can discuss the following:
# - Strongest relationship: If the correlation value is close to 1 or -1, it indicates a strong linear relationship.
# - Weakest relationship: If the correlation is close to 0, it indicates no or very weak linear relationship.
```

4. Happiness Disparity:

```python
import pandas as pd

# Load the South Asian and Middle Eastern datasets
south_asia_df = pd.read_csv('south_asian_countries.csv')
middle_east_df = pd.read_csv('middle_eastern_countries.csv')

# Calculate the Range and Coefficient of Variation for both regions

# For South Asia
south_asia_score = south_asia_df['score']
south_asia_range = south_asia_score.max() - south_asia_score.min()
south_asia_mean = south_asia_score.mean()
south_asia_std = south_asia_score.std()
south_asia_cv = (south_asia_std / south_asia_mean) * 100

# For Middle East
middle_east_score = middle_east_df['score']
middle_east_range = middle_east_score.max() - middle_east_score.min()
middle_east_mean = middle_east_score.mean()
middle_east_std = middle_east_score.std()
middle_east_cv = (middle_east_std / middle_east_mean) * 100

# Display the results
print(f"South Asia - Range: {south_asia_range}, CV: {south_asia_cv}%")
print(f"Middle East - Range: {middle_east_range}, CV: {middle_east_cv}%")

# Determine which region has greater variability
if south_asia_cv > middle_east_cv:
    print("South Asia has greater variability in happiness scores.")
else:
    print("Middle East has greater variability in happiness scores.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the South Asian and Middle Eastern datasets
south_asia_df = pd.read_csv('south_asian_countries.csv')
middle_east_df = pd.read_csv('middle_eastern_countries.csv')

# Function to identify outliers using the 1.5 * IQR rule
def identify_outliers(df, column1, column2):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1_1 = df[column1].quantile(0.25)
    Q3_1 = df[column1].quantile(0.75)
    IQR_1 = Q3_1 - Q1_1

    Q1_2 = df[column2].quantile(0.25)
    Q3_2 = df[column2].quantile(0.75)
    IQR_2 = Q3_2 - Q1_2

    # Define lower and upper bounds for outliers
    lower_bound_1 = Q1_1 - 1.5 * IQR_1
    upper_bound_1 = Q3_1 + 1.5 * IQR_1

    lower_bound_2 = Q1_2 - 1.5 * IQR_2
    upper_bound_2 = Q3_2 + 1.5 * IQR_2

    # Identify outliers in both columns
    outliers = df[(df[column1] < lower_bound_1) | (df[column1] > upper_bound_1) |
                  (df[column2] < lower_bound_2) | (df[column2] > upper_bound_2)]

    return outliers

# Identify outliers in South Asia and Middle East based on Score and GDP per Capita
south_asia_outliers = identify_outliers(south_asia_df, 'score', 'Log GDP per capita')
middle_east_outliers = identify_outliers(middle_east_df, 'score', 'Log GDP per capita')

# Visualize the outliers for South Asia
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, label='Normal')
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_outliers, color='red', label='Outliers')
plt.title('South Asia: Outliers based on GDP per Capita and Score')
plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.legend()
plt.show()

# Visualize the outliers for Middle East
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_df, label='Normal')
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_outliers, color='red', label='Outliers')
plt.title('Middle East: Outliers based on GDP per Capita and Score')
plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.legend()
plt.show()

# Display the outliers for further discussion
print("South Asia Outliers:\n", south_asia_outliers[['Country name', 'score', 'Log GDP per capita']])
print("\nMiddle East Outliers:\n", middle_east_outliers[['Country name', 'score', 'Log GDP per capita']])
```

7. Visualization:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the South Asian and Middle Eastern datasets
south_asia_df = pd.read_csv('south_asian_countries.csv')
middle_east_df = pd.read_csv('middle_eastern_countries.csv')

# Add a 'Region' column to each dataframe for identification
south_asia_df['Region'] = 'South Asia'
middle_east_df['Region'] = 'Middle East'

# Concatenate both DataFrames into one
combined_df = pd.concat([south_asia_df[['Country name', 'score', 'Region']],
                         middle_east_df[['Country name', 'score', 'Region']]], ignore_index=True)

# Create a boxplot comparing the distribution of Score between South Asia and the Middle East
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='score', data=combined_df)
plt.title('Distribution of Happiness Scores in South Asia vs Middle East')
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.show()
```
